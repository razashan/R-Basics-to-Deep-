# Linear Regression

Linear regression is used to predict the value of an outcome variable y on the basis of one or more input predictor variables x. In other words, linear regression is used to establish a linear relationship between the predictor and response variables.

In linear regression, predictor and response variables are related through an equation in which the exponent of both these variables is 1. Mathematically, a linear relationship denotes a straight line, when plotted as a graph.

There is the following general mathematical equation for linear regression:

```r
library("IRdisplay")
display_png(file="C.PNG")  
```

* y is a response variable.
* x is a predictor variable.
* a and b are constants that are called the coefficients.

**lm(formula,data)**

```r
x <- c(151, 174, 138, 186, 128, 136, 179, 163, 152, 131) ## independent 
y <- c(63,   81,  56,  91,  47,  57,  76,  72,  62,  48) ## dependent
```

```r
relation <- lm(y~x)
```

```r
print(relation)
```

```r
print(summary(relation))
```

# predict() Function

predict(object, newdata)
* object is the formula which is already created using the lm() function.

* newdata is the vector containing the new value for predictor variable.

```r
# The predictor vector.
x <- c(151, 174, 138, 186, 128, 136, 179, 163, 152, 131)

# The resposne vector.
y <- c(63, 81, 56, 91, 47, 57, 76, 72, 62, 48)
```

```r
relation <- lm(y~x)
# Find weight of a person with height 170.
a <- data.frame(x = 170)
result <-  predict(relation,a)
print(result)
```

abline() can be used to add vertical, horizontal or regression lines to a graph

cex. number indicating the amount by which plotting text and symbols should be scaled relative to the

pch is the shape of your data points

```r
plot(y,x,col = "blue",main = "Height & Weight Regression",
abline(lm(x~y)),cex = 1.3,pch = 11,xlab = "Weight in Kg",ylab = "Height in cm")
```

### Linearity
The relationship between the independent and dependent variable must be linear. We can test this visually with a scatter plot to see if the distribution of data points could be described with a straight line.

```r
plot(happiness ~ income, data = data)
```

```r
income_happiness_lm <- lm(happiness ~ income, data = data)

summary(income_happiness_lm)
```

```r
income.graph<-ggplot(data, aes(x=income, y=happiness))+
                     geom_point()
income.graph
```

Add the regression line using geom_smooth() and typing in lm as your method for creating the line. This will add the line of the linear regression

```r
income.graph <- income.graph + geom_smooth(method="lm", col="black")

income.graph
```

```r
library("ggpubr")
ggscatter(data, x = "income", y = "happiness", 
          add = "reg.line", conf.int = TRUE, 
          cor.coef = TRUE, cor.method = "pearson",
          xlab = "income", ylab = "happiness")
```
