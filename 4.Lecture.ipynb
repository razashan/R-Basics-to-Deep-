Reorder rows by Sepal.Length in descending order. Use the function desc():

```r
my_data %>% arrange(desc(Sepal.Length))
```

Instead of using the function **desc()**, you can prepend the sorting variable by a minus sign to indicate descending order, as follow.

# Transmute

transmute(): compute new columns but drop existing variables.

```r
my_data <- as_tibble(iris)
my_data
```

```r
my_data %>% 
  mutate(sepal_by_petal_l = Sepal.Length/Petal.Length)
```

```r
my_data %>%
  transmute(sepal_by_petal_l = Sepal.Length/Petal.Length,
    sepal_by_petal_w = Sepal.Width/Petal.Width)
```

# Group_by

```r
data <- read.csv("https://raw.githubusercontent.com/guru99-edu/R-Programming/master/lahman-batting.csv") 
```

```r
data %>%
group_by(lgID) %>%
summarise(mean_run = mean(HR))
```

```r
data
```

```r
library(ggplot2)
data %>% 
group_by(lgID) %>% 
summarise(mean_home_run = mean(HR)) %>% 

ggplot(aes(x = lgID, y = mean_home_run, fill = lgID)) +
    geom_bar(stat = "identity") +
    theme_classic() +
    labs(
        x = "baseball league",
        y = "Average home run",
        title = paste(
            "Example group_by() with summarise()"
        )
    )
```

```r
library(ggplot2)
data %>% 
group_by(lgID) %>% 
summarise(mean_home_run = mean(HR)) %>% 

ggplot(aes(x = lgID, y = mean_home_run, fill = lgID)) +
    geom_bar(stat = "identity") +
    theme_classic() +
    labs(
        x = "baseball league",
        y = "Average home run",
        title = paste(
            "Example group_by() with summarise()"
        )
    )
```
