# DATA CLEANING: 
To create country level and global combined data

Convert each data set from wide to long AND aggregate at country level

```r
library(tidyr)
```

```r
library(dplyr)
```

```r
confirmed <- confirmedraw %>%
gather(key ="date",value="confirmed",-c(Country.Region,Province.State,Lat,Long))%>%
group_by(Country.Region,date)%>%
summarize(confirmed = sum(confirmed))
```

```r
head(confirmed)
```

```r
deaths <- deathsraw %>%
gather(key="date",value="deaths",-c(Country.Region,Province.State,Lat,Long))%>%
group_by(Country.Region,date)%>%
summarize(deaths = sum(deaths))
```

```r
head(deaths)
```

```r
recovered <- recoveredraw %>%
gather(key="date",value="recovered",-c("Country.Region","Province.State","Lat","Long"))%>%
group_by(Country.Region,date)%>%
summarize(recovered = sum(recovered))
```

```r
head(recovered)
```

```r
summary(confirmed)
```

```r
summary(deaths)
```

```r
summary(recovered)
```

# Final data: combine all three

```r
country <- full_join(confirmed,deaths)%>% 
full_join(recovered)
```

```r
head(country)
```

# Fix date variable and convert from character to date

```r
str(country)
```

```r
country$date <- country$date %>%
sub("X","",.)%>%
as.Date("%m.%d.%y")
```

```r
str(country)
```

```r
head(country)
```

# Create new variable: number of days

```r
country <- country %>%
group_by(Country.Region)%>%
mutate(cumconfirmed = cumsum(confirmed), days = date - first(date)+1)
```

```r
head(country)
```

# Aggregate at world level

```r
world <- country %>%
group_by(date)%>%
summarize(confirmed = sum(confirmed),cumconfirmed = sum(cumconfirmed),
deaths = sum(deaths),recovered = sum(recovered))%>%
mutate(days = date - first(date)-1)
```

# Exploratory Data Analysis 

```r
library(ggplot2)
```

```r
ggplot(world,aes(x = date,y =confirmed))+
geom_bar(stat="identity",width=0.1)+
theme_classic()+
labs(title = "COVID-19 Global Confirmed Cases",x="Date",y="Confirmed Cases")+
theme(plot.title = element_text(hjust = 0.5))
```
