# Track COVID-19 Data

```r
library(lubridate)
library(tidyverse)
```

```r
data <- read.csv("https://opendata.ecdc.europa.eu/covid19/casedistribution/csv", na.strings = "", fileEncoding = "UTF-8-BOM", stringsAsFactors = F)
```

```r
data
```

```r
sum(data$cases)
```

```r
colnames(data)
```

 ### Total cases and max single day by country

```r
data %>% 
 group_by(countriesAndTerritories) %>% 
 summarise(cases_sum = sum(cases_weekly), cases_max = max(cases_weekly)) %>% 
 arrange(desc(cases_sum))%>%
 top_n(20)
```

### Total Deaths Worldwide

```r
sum(data$deaths_weekly)
```

### Total deaths and max single day by country

```r
data %>%
group_by(countriesAndTerritories)%>%
summarise(death_sum = sum(deaths_weekly), deaths_max = max(deaths_weekly))%>%
arrange(desc(death_sum))
```

# Plotting Daily Cases and Deaths

```r
us <- data[data$countriesAndTerritories == 'United_States_of_America',]
```

```r
us
```

```r
library(ggplot2)
```

```r
US_cases <- ggplot(us, 
 aes(dateRep, as.numeric(cases_weekly))) +
 geom_col(fill = "blue", alpha = 0.6) + 
 theme_minimal(base_size = 14) +
 xlab(NULL) + ylab(NULL) 
 
US_cases + labs(title="Daily COVID-19 Cases in US")
```

```r
US_deaths <- ggplot(us, 
 aes(dateRep, as.numeric(deaths_weekly))) +
 geom_col(fill = "purple", alpha = 0.6) + 
 theme_minimal(base_size = 14) +
 xlab(NULL) + ylab(NULL) 
 #scale_x_date(date_labels = "%Y/%m/%d")
US_deaths + labs(title="Daily COVID-19 Deaths in US")
```
