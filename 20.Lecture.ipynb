# skim() function

Like summary(), skim()'s method for data frames presents results for every column; the statistics it provides depend on the class of the variable.

Results of skim() are printed horizontally, with one section per variable type and one row per variable.

```r
skim(iris)
```

```r
skim(iris)%>%
dplyr::filter(skim_variable == "Petal.Length")
```

```r
iris %>%
skim()%>%
yank("numeric")
```

```r
iris %>%
skim()%>%
to_long()%>%
head()
```
