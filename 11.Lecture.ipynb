# R aggregate() function 

aggregate(data,by = list(data$column-name),FUN = mean)

R aggregate() function with mean

* mean()
* median()
* sum()
* sd()
* min()
* max()
* first()
* last()
* nth()
* n_distinct()

```r
values <- data.frame(value = c("a", "a", "a", "a", "a", 
                               "b", "b", "b", 
                               "c", "c", "c", "c"))
nr.of.appearances <- aggregate(x = values, 
                               by = list(unique.values = values$value), 
                               FUN = length)
```

```r
data <- read.csv("https://raw.githubusercontent.com/guru99-edu/R-Programming/master/lahman-batting.csv")%>%
select(c(playerID, yearID, AB, teamID, lgID, G, R, HR, SH))%>%
arrange(playerID, teamID, yearID)
```

```r
data
```

```r
summarise(data, mean_run =mean(R))
```

You can add as many variables as you want. You return the average games played and the average sacrifice hits.

```r
summarise(data, mean_games = mean(G),
    mean_SH = mean(SH, na.rm = TRUE))
```

```r
data %>%
group_by(lgID) %>%
summarise(mean_run = mean(HR))
```

```r
data %>%
group_by(lgID) %>%
summarise(mean_run = mean(HR),
         mean_games = mean(G))
```

```r
library(ggplot2)
data %>%
group_by(lgID)%>%
summarise(mean_home_run = mean(HR))%>%
ggplot(aes(x=lgID,y=mean_home_run,fill=lgID))+
geom_bar(stat="identity")+
theme_classic()+
labs(
x="Baseball league",
y= "Average Home Run",
title=paste("Example Group by Function with Summarise"))
```

**Compute the average number of games played by year.**

```r
ex1 <- data %>%
group_by(yearID) %>%
summarise(mean_game_year = mean(G))
head(ex1)
```

 **Show the summary statistic with a line plot and see the trend.**

```r
ggplot(ex1, aes(x = yearID, y = mean_game_year)) +
    geom_line() +
    theme_classic() +
    labs(
        x = "Year",
        y = "Average games played",
        title = paste(
            "Average games played from 1871 to 2016"
        )
    )
```

```r
data %>%
group_by(lgID) %>%
summarise(sum_homerun_league = sum(HR))
```

```r
data %>%
group_by(teamID) %>%
summarise(sd_at_bat_league = sd(HR))
```

```r
data %>%
group_by(playerID) %>%
summarise(min_G = min(G),
max_G = max(G))
```

```r
data %>%
group_by(playerID)%>%
summarise(number_year = n())%>%
arrange(desc(number_year))
```

```r
data %>%
group_by(playerID) %>%
summarise(first_appearance = first(yearID),
last_appearance = last(yearID))
```
